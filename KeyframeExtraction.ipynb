## Middleframe extraction ##
Extracting the most representative frames per scene, from the scene segmentation folder. 

Why stable shot analysis? 
-> This avoids transition frames at the beginning or end of a scene.

```python
import os
import cv2
import numpy as np
import time

# Function to check if OpenCV is installed
def check_opencv():
    if not cv2.__version__:
        print("‚ö†Ô∏è OpenCV is not installed. Please install it using 'pip install opencv-python'.")
        exit(1)


check_opencv()

# Function to safely get folder path from user
def get_folder_path(prompt, create_if_missing=False):
    while True:
        path = input(prompt).strip()
        if path:
            if os.path.exists(path) or create_if_missing:
                os.makedirs(path, exist_ok=True)  # Create folder if missing
                print(f"‚úÖ Using folder: {path}")
                return path
        print("‚ùå Invalid or empty path. Please enter a valid folder path.")


scene_video_base_folder = get_folder_path("Enter the base folder where segmented videos are stored (e.g., /Users/alicja/Desktop/VideoAnalyse/Corpus/VideoSegmentation): ")
keyframe_output_folder = get_folder_path("Enter the folder where all keyframes should be stored (e.g., /Users/alicja/Desktop/VideoAnalyse/Corpus/Keyframes): ", create_if_missing=True)

# Function to extract the middle frame from each scene
def extract_middle_keyframe(video_path, output_folder):
    print(f"Extracting keyframe from {video_path}...")
    video_filename = os.path.basename(video_path).split('.')[0]
    
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Check if the video is valid
    if frame_count == 0:
        print(f"Skipping {video_filename} (No frames detected)")
        cap.release()
        return

    middle_frame_idx = frame_count // 2  # Select middle frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, middle_frame_idx)
    success, frame = cap.read()

    if success:
        output_filename = os.path.join(output_folder, f"{video_filename}_keyframe.jpg")
        cv2.imwrite(output_filename, frame)
        print(f"‚úÖ Saved keyframe at {output_filename}")
    else:
        print(f"‚ùå Failed to extract keyframe from {video_filename}")
    
    cap.release()
    print(f"‚úÖ Keyframe extraction complete for {video_filename}")

# Function to manually select which segmented movie folders to process
def select_movies(movies):
    print("üìÇAvailable segmented movie folders:")
    for idx, movie in enumerate(movies, start=1):
        print(f"{idx}. {movie}")
    
    selected_indices = input("Enter the numbers of the movies you want to process (comma-separated, e.g., 1,3,5): ")
    selected_indices = [int(i.strip()) - 1 for i in selected_indices.split(',') if i.strip().isdigit() and 0 <= int(i.strip()) - 1 < len(movies)]
    
    if not selected_indices:
        print(" No valid selections made. Exiting.")
        exit()
    
    selected_movies = [movies[i] for i in selected_indices]
    return selected_movies

# Function to process videos in selected segmented movie folders
def process_selected_videos():
    segmented_movies = [movie for movie in os.listdir(scene_video_base_folder) if os.path.isdir(os.path.join(scene_video_base_folder, movie))]
    selected_movies = select_movies(segmented_movies)
    
    for movie in selected_movies:
        movie_path = os.path.join(scene_video_base_folder, movie)
        print(f" Processing segmented movie folder: {movie}")
        
        video_files = [f for f in os.listdir(movie_path) if f.endswith(".mp4")]
        
        if not video_files:
            print(f"No video files found in {movie_path}, skipping.")
            continue
        
        print(f"Processing all scene videos in: {movie}")
        for video in video_files:
            video_path = os.path.join(movie_path, video)
            extract_middle_keyframe(video_path, keyframe_output_folder)
        
        print(f"‚úÖ Finished processing all scenes in {movie}")


process_selected_videos()

print("‚úÖ Keyframe extraction process completed!")


```
