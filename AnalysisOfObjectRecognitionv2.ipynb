```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load object frequency dataset
file_path = "/Users/alicja/Desktop/VideoAnalyse/ObjectRecognition_Results/object_frequency_by_genre.csv"
df = pd.read_csv(file_path)

# Remove "person" from the dataset
df = df[df['class'] != "person"]


df_pivot = df.pivot_table(index='class', columns='genre', values='count', aggfunc='sum', fill_value=0)

# Assign genre to each object based on its highest frequency
object_genre_mapping = df_pivot.idxmax(axis=1)  # Finds the genre where each object appears most

# Log transform the data to balance frequent and rare objects
df_pivot = np.log1p(df_pivot)

# Standardize the data before PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_pivot)

# Apply PCA for 2D visualization
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'], index=df_pivot.index)

# Apply K-Means Clustering
optimal_k = 3  # Choosing 3 clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
pca_df['Cluster'] = kmeans.fit_predict(pca_data)

# Assign colors based on genre
genre_colors = {
    'Action': 'red',
    'Documentary': 'blue',
    'Drama': 'green'
}
pca_df['Genre'] = object_genre_mapping  
pca_df['Color'] = pca_df['Genre'].map(genre_colors) 

# Define marker styles for clusters
marker_styles = {0: 'o', 1: 's', 2: 'D'}  # Circle, square, and diamond markers for clusters

# Plot PCA visualization with genre-based colors and cluster-based markers
plt.figure(figsize=(12, 6))

for cluster in range(optimal_k):
    cluster_data = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(
        cluster_data['PC1'], cluster_data['PC2'],
        c=cluster_data['Color'], edgecolors="black", marker=marker_styles[cluster], label=f'Cluster {cluster + 1}'
    )

# Add object labels to key points
for i, obj in enumerate(pca_df.index):
    if i % 5 == 0:  # Reduce clutter by labeling every 5th object
        plt.text(pca_df.iloc[i, 0], pca_df.iloc[i, 1], obj, fontsize=8)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA & K-Means Clustering of Objects Across Genres (Colored by Genre, Clustered Separately)")

# Create a legend for clusters and genres
handles = [plt.Line2D([0], [0], marker=marker_styles[i], color='w', markersize=10, markeredgecolor='black',
                       markerfacecolor='gray', label=f'Cluster {i + 1}') for i in range(optimal_k)]
genre_legend = [plt.Line2D([0], [0], marker='o', color='w', markersize=10, markeredgecolor='black',
                            markerfacecolor=color, label=genre) for genre, color in genre_colors.items()]

plt.legend(handles=handles + genre_legend, title="Legend", loc="upper right")
plt.show()


```
